## 3a - Normalizing values

```python
def normalize(val,min,max):
  return (val-min)/(max-min)
normalized_cols = []
for col in df2.columns[2:7]:
  min = df2[col].min()
  max = df2[col].max()
  df2[col]=df2[col].apply(normalize,args=[min,max])
  print("normalized: ",col)
  normalized_cols.append(col)
```

```python
df2.loc[:,normalized_cols]
```
